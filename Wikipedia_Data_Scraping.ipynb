```python
import requests
from bs4 import BeautifulSoup
```

```python
name = str(input()).replace(' ','+')
google_link = 'https://www.google.com/search?q='+name+'+wikipedia'
print(google_link)
```

```python
res = requests.get(google_link)
soup1 = BeautifulSoup(res.text,'html.parser')
print(soup1)
```

```python
for sp in soup1.find_all('div'):
  try:
    link = sp.find('a').get('href')
    if('en.wikipedia.org' in link):
      break
  except:
    pass
link = link[7:].split('&')[0]
print(link)
```

```python
res1 = requests.get(link)
soup2 = BeautifulSoup(res1.text,'html.parser')
```

```python
para = ''
for p in soup2.find_all('p'):
  para += p.text
  para += '\n'
print(para)
```

```python
para = para.strip()
```

```python
# Removing Refrence Numbers (like [2])

for i in range(3,132):
  a = '[' + str(i) + ']'
  para = para.replace(a,' ')
print(para)
```

```python
heading = soup2.find('h1').text
print(heading)
```
